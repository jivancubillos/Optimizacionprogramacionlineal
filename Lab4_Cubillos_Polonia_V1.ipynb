```python
#Punto A

import itertools as it
import numpy as np
import time
from scipy.optimize import linprog
A1=[[1,1,1,1,1],[1,2,3,4,5],[4,3,3,3,4]]
A2=[[1,1,1,1,1],[1,2,3,4,5],[2,3,4,5,6]]
A1=np.array(A1)
A2=np.array(A2)
B=[1,2,3]
def solucionesbasicas(A,B):
  lenb = len(B)
  solucionesbasica = []
  i = 0
  m = len(A)
  n = len(A[0])
  comb = it.combinations(range(n),lenb)
  comb = list(comb)
  d = len(comb)
  i = 0
  base=[]
  MA=[]
  while i<d:
    Matriz = []
    j = 0
    k = 0
    while k<lenb:
      Matriz.append([])
      k+=1
    while j<lenb:
      a = comb[i][j]
      k = 0
      while k<lenb:
        Matriz[k].append(A[k][a])
        k+= 1
      j+=1
      MA.append(list(A[:,a]))
    MA=np.transpose(MA)
    determinante=np.linalg.det(Matriz)
    if determinante!=0:
      sol = np.linalg.solve(np.array(Matriz),np.array(B))
      solucionesbasica.append(list(sol))
    else:
      solucionesbasica.append('La Matriz base es singular')
    base.append(list(MA))
    MA=[]
    i+=1
  dir={'Solucion': solucionesbasica,'Base': base}
  return dir
```

```python
#Punto B y C

import pandas as pd
from IPython.display import display

pd.set_option('max_colwidth',75)

SolBasRC=solucionesbasicas(A1,B)
SolBasRD=solucionesbasicas(A2,B)

df_SolBasRC=pd.DataFrame(SolBasRC)
print('\n Matriz de rango completo: \n')
display(df_SolBasRC)

print('\n Matriz de rango deficiente): \n')
display(pd.DataFrame(SolBasRD))
```

# Segundo Punto

a)
Una planta de energía quema carbón, petróleo y gas para generar electricidad. Suponga que cada
tonelada de carbón genera 600 kilovatios hora, emite 20 unidades de bióxido de azufre y 15 unidades
de partículas suspendidas y tiene un costo de 200 dólares; cada tonelada de petróleo genera 550
kilovatios hora, emite 18 unidades de bioxido de azufre y 12 unidades de partículas suspendidas y tiene
un costo de 220 dólares; cada tonelada de gas genera 500 kilovatios hora, emite 15 unidades de bióxido de
azufre y 10 unidades de partículas suspendidas y tiene un costo de 250 dólares.
La oficina de protección ambiental restringe las emisiones diarias de bióxido de azufre a no más de 60
unidades y las partículas suspendidas a no más de 75. Si la planta de energía tiene un presupuesto para
este proyecto de 2000 dólares diarios en combustible, ¿cuánto combustible de cada clase debe comprar
para maximizar la cantidad de energía generada?

Variables de desición
$x_c$: Toneladas de carbón
$x_p$: Toneladas de petróleo
$x_g$: Toneladas de gas

Función de costo, se busca maximizar la cantidad de energía generada.

$\max  600x_c+550x_p+500x_g$ (kWh)

Restricciones del problema:

Limita la cantidad de unidades de bióxido de azufre producida por la planta de energía diariamente.

$20x_c+18x_p+15x_g\leqq 60$ (unidades de bióxido de azufre)

Limita la cantidad de unidades de partículas suspendidas producidas por la planta de energía diariamente.

$15x_c+12x_p+10x_g\leqq 75$ (unidades de partículas suspendidas)

Limita el dinero utilizado por la planta de energía diariamente.

$200x_c+220x_p+250x_g\leqq 2000$ ($)

Las variables $x_c,x_p,x_g\geqslant0$

Problema de programación lineal en forma estándar

$\min -600x_c-550x_p-500x_g$

s.a.

$20x_c+18x_p+15x_g+s_1 = 60$
$15x_c+12x_p+10x_g+s_2 = 75$
$200x_c+220x_p+250x_g+s_3 = 2000$

$x_c,x_p,x_g,s_1,s_2,s_3\geqslant0$

Problema de la forma $Ax\leqq b$ 

$\min \begin{bmatrix}
-600\\ -550\\ -500 \\
\end{bmatrix}^Tx$


$\begin{bmatrix}
20 & 18 & 15 \\
15 & 12 & 10 \\
200 & 220 & 250 \\
\end{bmatrix}x \leqq
\begin{bmatrix}
60 \\
75 \\
2000 \\
\end{bmatrix}$

```python
# subpuntos B
A=[[20,18,15],[15,12,10],[200,220,250]]
b=[60,75,2000]
c=[-600,-550,-500]
A=np.array(A)
b=np.array(b)
def Puntob(A,b):
  sol = solucionesbasicas(A,b)
  solbasicas = sol['Solucion']
  costos = []
  tipo = []
  for i in solbasicas:
    es_factible = 1;
    j = 0
    while j<len(i) and es_factible == 0:
      if i[j]<0:
        es_factible = 0
        tipo.append("Básica")
      j += 1
    if es_factible == 1:
      tipo.append("Básica factible")
    costo = np.dot(np.array(c),np.array(i))
    costos.append(costo)
  m=[sol,tipo,costos]
  return m
solucion=Puntob(A,b)[0]
tipo=Puntob(A,b)[1]
costos=Puntob(A,b)[2]

TablaSolucion = pd.DataFrame(solucion)
TablaSolucion["Tipo de solución"] = tipo
TablaSolucion["Valor"] = costos

display(TablaSolucion)
```
